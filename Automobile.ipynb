```python
#check for missing value in the datasets
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Plot histogram of 'horsepower' column to examine its distribution
plt.hist(df['horsepower'].dropna(), bins=30, alpha=0.5, color='red', edgecolor='black')
plt.title('Horsepower Distribution')
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#different cars and their origin
df.groupby('name')['origin'].unique()
```

```python
cylinder_no = df['cylinders'].unique()
horsepower_means = []
for n in cylinder_no:
    cylinder = df[df['cylinders']==n]
    horsepower_mean = cylinder['horsepower'].mean()
    horsepower_means.append(horsepower_mean)

plt.bar(cylinder_no, horsepower_means)
plt.xlabel('Number of cylinders')
plt.ylabel('Mean horsepower')
plt.show()
```

Observation:
 It is quite obvious that more number of cylinder mean more strength. But the interesting thing here is that even if there are only 3 cylinders in some cars, its power is quite higher than the cars having 3 and 4 cylinders and somehow on-par with cars having 5 cylinders.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

df.groupby('origin')['weight'].mean().plot(kind='bar', ax=axs[0])
axs[0].set_title('Mean Weight by Origin')
#axs[0].set_xlabel('Origin')
axs[0].set_ylabel('Weight')

df.groupby('origin')['acceleration'].mean().plot(kind='bar', ax=axs[1])
axs[1].set_title('Mean Acceleration by Origin')
#axs[1].set_xlabel('Origin')
axs[1].set_ylabel('Acceleration')

plt.show()
```

Observation: It can be seen that the weight of American cars are much heavier than cars from Japan and Europe. As a result of increased mass, the acceleration is less than heavier cars.
