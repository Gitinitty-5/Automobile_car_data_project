```python
# Create visualizations
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(df,x='Store_Area', y='Store_Sales')
plt.title('Store Area by Store Sales')
plt.show()
```

```python
#plotting available items and store area on scatter plot. Graph shows positive correlation as expected the bigger the store,
# the more items
fig = px.scatter(df, x="Items_Available", y="Store_Area",
              title="Store area vs Items Available",
              trendline="ols",
               trendline_color_override="red",
              template="plotly_dark")
fig.show()
```

```python
#plotting available items and store area on scatter plot. Graph shows positive correlation as expected the bigger the store,
# the more items
fig=px.scatter(df,
              x="Items_Available",
              y="Store_Area",
              title="Store area by Items Available and Store Sales",
              trendline="ols",
               color="Store_Sales",
               size="Store_Sales",
               size_max=15,
               trendline_color_override="red",
              template="plotly_dark")
fig.show()
```
